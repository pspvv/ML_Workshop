```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

###1. Basic Line Plot
o Plot a simple line graph of ùë¶ = ùë•^2 for ùë• values ranging from -10 to 10.


o Label the x-axis as "X Values" and the y-axis as "Squared Values."

o Add a title to the graph.

```python
x = np.arange(-10,11)# Generating values from -10 to 10
y = x**2
plt.plot(x,y) #Plotting a basic line plot
plt.xlabel("X Values")
plt.ylabel("Squared Values")
plt.title("Line Plot of y = x^2")
plt.show()
```

###2. Bar Plot
o Create a bar chart showing the marks scored by 5 students:
['Sai', 'Vamsi',
'Divij', 'Sritan', 'Sita'] with marks [85, 90, 78, 92, 88].

o Add labels to the x-axis and y-axis and give the chart a title.


```python
plt.bar(['Sai', 'Vamsi', 'Divij', 'Sritan', 'Sita'],[85, 90, 78, 92, 88]) #Creating a bar chart with the given data and adding the labels to them
plt.title('Marks Scored by Students')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.show()
```

###3. Scatter Plot
o Generate two lists of 20 random integers each (for x and y coordinates).


```python
# Generating two lists of 20 random integers
x = np.random.randint(0,100,20)
y = np.random.randint(0,100,20)
print("Random values")
x,y
```

o Create a scatter plot with these values, and customize it by adding color,
size, and labels to the points.

```python
plt.scatter(x,y,color='red',s=20,marker="^",label="Data points") # Creating a scatter plot by customising the colour size and adding label to points
plt.xlabel('X Coordinates')
plt.ylabel('Y Coordinates')
plt.title("Scatter plot")
plt.legend()
plt.show()
```

###4. Histogram
o Simulate 100 random values from a normal distribution using
np.random.randn() and plot a histogram.



```python
data = np.random.randn(100) #Generating 100 random values from a normal distribution
print("100 random values")
print(data)
```

o Customize the number of bins and add a grid to the plot.

```python
plt.hist(data, bins=20, edgecolor='black')# customising the number of bins to be in the plot
plt.grid(True)# Setting up the grid in the plot
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram of Random Data")
plt.show()
```

###5. Pie Chart
o Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E']
with values [20, 30, 25, 15, 10].


```python
categ=['A', 'B', 'C', 'D', 'E']
val= [20, 30, 25, 15, 10]
print(categ)
print(val)
```

o Display the percentage values on the chart and use a color palette for
differentiation.

```python
plt.pie(val, labels=categ, autopct='%1.1f%%', colors=['red', 'blue', 'green', 'orange', 'purple'])# Displaying the percentage values on the pie chart and using a colour palette
plt.title("Pie Chart of Categories")
plt.show()
```

###6. Box Plot (Seaborn)
o Create a box plot to visualize the distribution of the marks scored by
students in a class: [78, 85, 92, 88, 76, 95, 89, 84, 91, 73].

o Highlight the median and any outliers.

```python
marks = [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]
# Creating the box plot
sns.boxplot(marks,flierprops=dict(marker='o', color='yellow', markersize=25),
            medianprops=dict(color='red', linewidth=2))
plt.xlabel("Marks")
plt.title("Box Plot of Student Marks")
# Displaying the plot
plt.show()
```

###7. Heatmap (Seaborn)
o Create a 5x5 matrix with random integers between 1 and 10 using
np.random.randint() and visualize it as a heatmap using Seaborn.

o Add a color bar and annotate the heatmap with the values.


```python
data = np.random.randint(1, 11, size=(5, 5)) # Creating a matrix of 5x5 With random integers between 1 and 10
sns.heatmap(data, annot=True, cmap="YlOrRd", fmt="d", linewidths=.5) # Adding colours and annotating the heat map
plt.title("Heatmap")
plt.show()
```

###8. Comparison of Matplotlib and Seaborn


```python
data = {
    'Category': ['A', 'B', 'C'],
    'Values': [10, 20, 15]
  }
df = pd.DataFrame(data)
print(df)
```

o Generate a pair of plots (bar charts or line plots) using the same data in
both Matplotlib and Seaborn.

o Highlight the differences in the code, syntax, and aesthetics of the two
libraries.

```python
# Matplotlib
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.bar(df['Category'], df['Values'], color='blue')
plt.title("Matplotlib Bar Plot")

# Seaborn
plt.subplot(1, 2, 2)
sns.barplot(x='Category', y='Values', data=df, color='green')
plt.title("Seaborn Bar Plot")

# Displaying both plots made by matplotlib on the Seaborn
plt.tight_layout()
plt.show()
```

###9. Pair Plot (Seaborn)
